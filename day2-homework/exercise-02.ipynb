## Load packages

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Load data

```python
df_SNP = pd.read_csv ("matrix_1kg.txt",sep='\t')
df_SNP
```

## Calculate alternate allele frequency

```python
#calculate the counts of alternate alleles
alt_counts=df_SNP.iloc[:,4:].sum(axis=1)
alt_counts
```

```python
#calculate the frequency
sample_size=len(df_SNP.iloc[1,4:])
frequency=alt_counts/sample_size
frequency
```

## Plotting

```python
fig,axis = plt.subplots()
plt.hist(frequency)
plt.title('SNP frequency spectrum')
plt.xlabel('Allele frequency')
plt.ylabel('Count')
plt.show()
```
