# Load packages


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Load K4me3.bed K9me3.bed and K27me3.bed


```python
bed_columns = ["chr", "start", "end"]
df_k4 = pd.read_table( "../../qbb2020/data/K4me3.bed", names=bed_columns )
df_k4
```

```python
bed_columns = ["chr", "start", "end"]
df_k9 = pd.read_table( "../../qbb2020/data/K9me3.bed", names=bed_columns )
df_k9
```

```python
bed_columns = ["chr", "start", "end"]
df_k27 = pd.read_table( "../../qbb2020/data/K27me3.bed", names=bed_columns )
df_k27
```

# Calculate width

```python
df_k4.loc[:,'width']=df_k4.loc[:,'end']-df_k4.loc[:,'start']
df_k4
```

```python
df_k9.loc[:,'width']=df_k9.loc[:,'end']-df_k9.loc[:,'start']
df_k9
```

```python
df_k27.loc[:,'width']=df_k27.loc[:,'end']-df_k27.loc[:,'start']
df_k27
```

# Extract width
#### It is better to change the name of the data set after extracting the width like df_k4_width, or the original variable df_k4 will be changed.

```python
df_k4_width=df_k4.loc[:,"width"]
df_k4_width
```

```python
df_k9_width=df_k9.loc[:,"width"]
df_k9_width
```

```python
df_k27_width=df_k27.loc[:,"width"]
df_k27_width
```

```python
fig,ax=plt.subplots(nrows=3,figsize=(7,10))
ax[0].hist(df_k4_width)
ax[0].set_xlabel("width")
ax[0].set_ylabel("number of k4")
ax[0].set_title("K4 width distribution")

ax[1].hist(df_k9_width)
ax[1].set_xlabel("width")
ax[1].set_ylabel("number of k9")
ax[1].set_title("K9 width distribution")

ax[2].hist(df_k27_width)
ax[2].set_xlabel("width")
ax[2].set_ylabel("number of k27")
ax[2].set_title("K27 width distribution")
plt.tight_layout()
plt.show()
```
