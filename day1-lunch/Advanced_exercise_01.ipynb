# Load packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Load fpkms.csv

```python
df_fpkms = pd.read_csv( "../../qbb2020/data/fpkms.csv", index_col="t_name" )
df_fpkms
```

# Subset transcriptÂ¶


```python
goi = "FBtr0331261"
df_sxl = df_fpkms.loc[goi,:]
df_sxl
```

## Male subset

```python
df_sxl_male=df_sxl[1:9]
df_sxl_male
```

## Female subset

```python
df_sxl_female=df_sxl[9:]
df_sxl_female
```

## Plot male and female data in the same plot

```python
fig,ax=plt.subplots()
x=["10","11","12","13","14A","14B","14C","14D"]
plt.plot(x,df_sxl_male,label='male')
plt.plot(x,df_sxl_female,label='female')
plt.title('time course of FBtr0331261')
plt.legend()
plt.ylabel('fpkms')
plt.xlabel('stage')
plt.show()
```
