```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
x2 = [2, 4, 6]
y2 = [8, 64, 216]
```

## Set xlim and ylim the same for both subplots 

```python
fig, ax = plt.subplots( nrows=2 )
ax[0].plot( x, y, '.',label='x^2' )
ax[0].set_xlim(0.5,6.5)
ax[0].set_ylim(-5,230)
ax[0].set_title('Quadratic Polynomial')
ax[0].set_xlabel( "x" )
ax[0].set_ylabel( "f(x)" )
ax[0].legend()


ax[1].plot( x2, y2 ,label='x^3')
ax[1].set_xlim(0.5,6.5)
ax[1].set_ylim(-5,230)
ax[1].set_title('Cubic Polynomial')
ax[1].set_xlabel( "x" )
ax[1].set_ylabel( "f(x)" )
ax[1].legend()

plt.tight_layout()
plt.show()
```

### For each subplot in the figure, we have to add axis and labels separately with prefix ax[0]. or ax[1].
### Add plt. tight_layout could make two plots to the better position so there's no overlap of the titles and axis labels.
