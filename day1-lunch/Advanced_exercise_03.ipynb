## Add replication

```python
fig,ax=plt.subplots()
x=["10","11","12","13","14A","14B","14C","14D"]
y1_male=[]
y1_female=[]
y2_male=[]
y2_female=[]
for i in range(len(x)):
    y1_male.append(np.log2(df_sxl_male[i]+1))
    y1_female.append(np.log2(df_sxl_female[i]+1))
plt.plot(x,y1_male,'.',label='male1',color='blue')
plt.plot(x,y1_female,'.',label='female1',color='red')

for i in range(4):
    y2_male.append(np.log2(df_sxl_rep[i]+1))
    y2_female.append(np.log2(df_sxl_rep[4+i]+1))
    

plt.plot(x[4:],y2_male,'.',color='darkblue',label='male2')
plt.plot(x[4:],y2_female,'.',color='darkred',label='female2')

plt.title('log2 time course of FBtr0331261')
plt.legend()
plt.ylabel('log2(sxl)')
plt.xlabel('stage')
plt.show()
```
