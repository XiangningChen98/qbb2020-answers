```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
bed_columns = ["chr", "start", "end"]
df_k4 = pd.read_table( "../../qbb2020/data/K4me3.bed", names=bed_columns )
df_k4
```

```python
bed_columns = ["chr", "start", "end"]
df_k9 = pd.read_table( "../../qbb2020/data/K9me3.bed", names=bed_columns )
df_k9
```

```python
bed_columns = ["chr", "start", "end"]
df_k27 = pd.read_table( "../../qbb2020/data/K27me3.bed", names=bed_columns )
df_k27
```

```python
df_k4.loc[:,'width']=df_k4.loc[:,'end']-df_k4.loc[:,'start']
df_k9.loc[:,'width']=df_k9.loc[:,'end']-df_k9.loc[:,'start']
df_k27.loc[:,'width']=df_k27.loc[:,'end']-df_k27.loc[:,'start']
```

```python
fig,ax=plt.subplots()

y_df_k4 = np.log2(df_k4.loc[:,'width'])
y_df_k9 = np.log2(df_k9.loc[:,'width'])
y_df_k27 = np.log2(df_k27.loc[:,'width'])
label=['k4','k9','k27']
data_name=[y_df_k4,y_df_k9,y_df_k27]
for i in range(len(data_name)):
    ax.hist(data_name[i],label=label[i],bins=10,alpha=0.5,density=True,histtype='bar')
plt.legend()
plt.ylabel('density')
plt.xlabel('width')
plt.title('density plot of log2(width) distribution')
plt.show()
```

```python
fig,ax=plt.subplots()
y_df_k4 = np.log2(df_k4.loc[:,'width'])
y_df_k9 = np.log2(df_k9.loc[:,'width'])
y_df_k27 = np.log2(df_k27.loc[:,'width'])
label=['k4','k9','k27']
data_name=[y_df_k4,y_df_k9,y_df_k27]
for i in range(len(data_name)):
    ax.hist(data_name[i],label=label[i],bins=10,alpha=0.5,density=True,cumulative=True,histtype='bar')
plt.legend()
plt.ylabel('density')
plt.xlabel('width')
plt.title('cumulative of density plot of log2(width) distribution')
plt.show()
```

## Using log2 and density = True could make the y range similar for all of the three dataset, which is easier to observe.
