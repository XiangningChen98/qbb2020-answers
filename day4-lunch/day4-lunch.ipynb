## AE4:  Find the nearest protein coding gene for each position in a given list of positions
### build a function to make the process easier

```python
def find_nearest(chrom,position):
    df_gene=df_gtf[(df_gtf['seqname']==chrom)&(df_gtf['feature']=='gene')]
    df_3R_protein_coding=df_gene[ df_gene['attribute'].str.contains('protein_coding')]
    df_3R_cut=copy.deepcopy(df_3R_protein_coding)
    count=0
    for i in range(1):
        while len(df_3R_cut)>2: 
            count+=1
            if len(df_3R_cut)%2 == 0:
                cut_step=int(len(df_3R_cut)/2)-1
            else:
                cut_step=int(len(df_3R_cut)/2)       
            if df_3R_cut.iloc[cut_step,3]>position:
                df_3R_cut=df_3R_cut[:cut_step+1]
            else:
                df_3R_cut=df_3R_cut[cut_step:]
        while len(df_3R_cut)==2:
            count+=1
#             print(df_3R_cut)
            dis1=abs(df_3R_cut.iloc[0,3]-position)
            dis2=abs(df_3R_cut.iloc[1,3]-position)
            if dis1>dis2:
                df_3R_cut=df_3R_cut.iloc[[1]]
            else:
                df_3R_cut=df_3R_cut.iloc[[0]]
    attribute = list(df_3R_cut['attribute'].str.split(';'))
    df_3R_cut.insert(5,'gene_name',attribute[0][2][10:])
    df_3R_cut.insert(6,'gene_biotype',attribute[0][4][13:])
    df_3R_cut.insert(7,'count',count)
    
    return count,df_3R_cut

      
```

```python
restult_df_all= pd.DataFrame(columns =['seqname','source','feature','start','end','gene_name','gene_biotype','count','score','strand','frame','attribute'] )

fs=open("day3_100_positions.txt",'r')
for line in fs:
    pos=line.rstrip().split()
    
    count,restult_df=find_nearest(pos[0],int(pos[1]))
    restult_df_all=pd.concat([restult_df_all,restult_df])
restult_df_all
```
