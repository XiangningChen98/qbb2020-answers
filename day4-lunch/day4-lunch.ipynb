## Import gtf file

```python
df_gtf = pd.read_csv("BDGP6.Ensembl.81.gtf",skiprows = 5,sep = "\t",
                     names=['seqname','source','feature','start','end','score','strand','frame','attribute'])
df_gtf
```

## E1 - 1: Find and report the nearest protein coding gene to the position 3R:21,378,950.

```python
df_gtf.iloc[10,-1]
```

```python
#extract data on 3R and is gene
# 
df_3R_gene=df_gtf[(df_gtf['seqname']=='3R')&
             (df_gtf['feature']=='gene')
            ]
df_3R_gene

```

```python
#extract protein coding gene
df_3R_protein_coding=df_3R_gene[ df_3R_gene['attribute'].str.contains('protein_coding')]
df_3R_protein_coding
```

```python
df_3R_cut=copy.deepcopy(df_3R_protein_coding)
count=0
for i in range(15):
    
    while len(df_3R_cut)>1:
        count+=1
        cut_step=int(len(df_3R_cut)/2)
        if df_3R_cut.iloc[cut_step-1,3]>21378950:
            df_3R_cut=df_3R_cut[:cut_step]
        else:
            df_3R_cut=df_3R_cut[cut_step:]
print('Nearest protein coding gene to the position 3R:21,378,950 is')
      
df_3R_cut
            
            
    
    
```

## E1 - 2: Report the gene's linear genomic distance from 3R:21,378,950.

```python
distance_start = df_3R_cut['start']-21378950
distance_end = df_3R_cut['end']-21378950
print("The gene's start position is ", int(distance_start), "base from 3R:21,378,950")
print("The gene's end position is ", int(distance_end), "base from 3R:21,378,950")
```
