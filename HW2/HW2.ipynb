```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import io
from matplotlib import gridspec
```

```python
def read_vcf(path):
    with open(path, 'r') as f:
        lines = [l for l in f if not l.startswith('##')]
    return pd.read_csv(
        io.StringIO(''.join(lines)),
        dtype={'#CHROM': str, 'POS': int, 'ID': str, 'REF': str, 'ALT': str,
               'QUAL': str, 'FILTER': str, 'INFO': str},
        sep='\t'
    ).rename(columns={'#CHROM': 'CHROM'})

data=read_vcf("/Users/xiangning/qbb2020-answers/HW2/variant_calling/var_A01_all.ann.vcf")
data
### This code for reading vcf file was found on line: https://gist.github.com/dceoy/99d976a2c01e7f0ba1c813778f9db744
```

```python
## Read depth: DP
fig = plt.figure(figsize=(20, 8))
spec = gridspec.GridSpec(ncols=4, nrows=2,
                         width_ratios=[1,1,1,1])

read_depth=[]
for sample in ['09','11','23','24','27','31','35', '39', '62', '63']:

    
    info=list(data[sample])
    for item in info:
        item_sep=item.split(':')
        if item_sep[2]!='.':
            read_depth.append(int(item_sep[2]) )
ax0 = fig.add_subplot(spec[0,0])
ax0.hist(read_depth,log=True,bins=25)
ax0.set_xlabel('Read depth')
ax0.set_ylabel('Number')
ax0.set_title('Distribution of read depth')
    
    
## Quality: QA
quality=[]
for sample in ['09','11','23','24','27','31','35', '39', '62', '63']:
    info=list(data[sample])
    for item in info:
        item_sep=item.split(':')
#         print(item_sep)
        if item_sep[7]!='.':
            if ',' in item_sep[7]:
#                 print('yes',item_sep[7])
                string=item_sep[7].split(',')
                for i in string:
                    quality.append(int(i))
            else:
                quality.append(int(item_sep[7]) )
ax1 = fig.add_subplot(spec[0,1])
   
ax1.hist(quality,log=True,bins=25,color='green',alpha=0.6)
ax1.set_xlabel('Quality')
ax1.set_ylabel('Number')
ax1.set_title('Distribution of quality')
   
        
## Allele frequency: AF   

AF=[]
infos=list(data['INFO'])
for info in infos:
    items_sep=info.split(';')
    for item in items_sep:
        if ('AF=' in item) and ('SAF=' not in item):
            AF_info=item.split('=')
            if ',' in AF_info[1]:
#               
                string=AF_info[1].split(',')
                for i in string:
                    AF.append(float(i))
            
            else:
                AF.append(float(AF_info[1]))
ax2 = fig.add_subplot(spec[0,2])

ax2.hist(AF,log=True,bins=25,color='red',alpha=0.6)
ax2.set_xlabel('Allele frequency')
ax2.set_ylabel('Number')
ax2.set_title('Distribution of allele frequency')

## unction effect
ANN=[]
infos=list(data['INFO'])
for info in infos:
    items_sep=info.split(';')
    for item in items_sep:
        if 'ANN=' in item:  
            
            effectsinfo=item.split(',')
            for effectinfo in effectsinfo:
                effects=effectinfo.split('|')[1]
                ANN.append(effects)
ANN_dict={}
for ann in ANN:
    ANN_dict.setdefault(ann,0)
    ANN_dict[ann]+=1
ax3 = fig.add_subplot(spec[1,0:3])
for k,v in list(ANN_dict.items()):
    ax3.bar(k,v,log=True)
ax3.set_ylabel('Number')
ax3.tick_params(labelrotation=-90)
ax3.set_title('Function effects')
print(ANN_dict)
plt.tight_layout()
plt.savefig("step8_figure.png")
```
